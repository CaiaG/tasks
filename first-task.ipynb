```python
import wandb

# start a new wandb run to track this script
wandb.init(
   # set the wandb project where this run will be logged
   
   project="my-awesome-project",
    
   # track hyperparameters and run metadata
   config={
    "learning_rate": 0.02,
    "architecture": "DCNN",
    "dataset": "CIFAR-10",
    "epochs": 3,
    }
)

def objective(trial): 
    learning_rate = trial.suggest_float("learning_rate", 1e-5, 1e-1, log=True)
    dropout_rate = trial.suggest_float("dropout_rate", 0.0, 0.5)
    
    wandb.config.learning_rate = learning_rate
    wandb.config.dropout_rate = dropout_rate

    # get dset
    path = untar_data(URLs.CIFAR)

    # trainset.train and trainset.test & normalize pixel vals
    trainset = ImageDataLoaders.from_folder(path, train='train', valid_pct=0.2, test='test', batch_tfms=Normalize.from_stats(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5]))

    # make model - contains activation functions and pooling layers
    model = vision_learner(trainset, resnet18, metrics=accuracy)
    
    # train
    model.fine_tune(3, base_lr = learning_rate)

    # accuracy_value = model.validate()[1].item()

    _, accuracy_value = model.validate()

    return accuracy_value

study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=10) 

best_learning_rate = study.best_params["learning_rate"]
best_dropout_rate = study.best_params["dropout rate"]
wandb.config.learning_rate = best_learning_rate
wandb.config.dropout_rate = best_dropout_rate

wandb.finish()
# LJANKJDHKJBA
```
