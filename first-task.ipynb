```python
import wandb

# start a new wandb run to track this script
wandb.init(
   # set the wandb project where this run will be logged
   
   project="my-awesome-project",
    
   # track hyperparameters and run metadata
   config={
    "learning_rate": 0.02,
    "architecture": "DCNN",
    "dataset": "CIFAR-10",
    "epochs": 7,
    }
)

# get dset
path = untar_data(URLs.CIFAR)

# trainset.train and trainset.test & normalize pixel vals
trainset = ImageDataLoaders.from_folder(path, train='train', valid_pct=0.2, test='test', batch_tfms=Normalize.from_stats(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5]))

# make model - contains activation functions and pooling layers
model = vision_learner(trainset, resnet18, metrics=accuracy)

# train
model.fine_tune(7)

wandb.finish()
```
